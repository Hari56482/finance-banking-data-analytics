```python
import pandas as pd
df = pd.read_csv("../data/loan_customers.csv")
df.head()
```

I am checking the dataset's stucture and basic statistics to understand data types,missing values, and general distribution.

```python
df = df.rename(columns={
    'coustomer_id': 'customer_id',
    'laon_to_income': 'loan_to_income'
})
```

```python
df.describe().round(2)
```

"income has a wide range, which means customers have different financial backgrounds."

I am creating simple new features to help identify financial stress, such as loan-to-income ratio and a rule-based high-risk flag.

```python
df = df[df['income'] > 0].copy()
```

```python
df['loan_to_income'] = (df['loan_amount'] / (df['income'] + 1)).round(3)
df['balance_to_income'] = (df['current_balance'] / (df['income'] + 1)).round(3)
```

```python
df['high_risk_rule'] = (
    (df['credit_score'] <580) |
    (df['late_payments'] >= 2) |
    (df['loan_to_income'] > 0.6)
).astype(int)
df[['customer_id','credit_score','late_payments','loan_to_income','high_risk_rule']].head()
```

```python
df.columns.tolist()
```

I am now loading the cleaned dataframe into an in-memory SQLite database so i can run SQL queries for risk segmentation. This helps demonstrate SQL skills commonly used in credit risk analysis.

```python
import sqlite3
conn = sqlite3.connect(":memory:")
df.to_sql("customers", conn, index=False, if_exists="replace")
```

```python
query = """
SELECT
(credit_score / 50) * 50 AS score_bucket,
COUNT(*) AS total_customers,
SUM(loan_status) AS total_defaults,
ROUND(100.0 * SUM(loan_status) / COUNT(*), 2)AS default_rate
FROM customers
GROUP BY score_bucket
ORDER BY score_bucket;
"""
```

```python
pd.read_sql(query, conn)
```

The sql analysis shows thatvthe customers with lower credit scores have higher default rates .the 500-550 bucket is significantly riskier than the higher score that ranges in the analysis.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
features = ['credit_score', 'late_payments', 'loan_to_income']
target = 'loan_status'

x = df[features]
y = df[target]
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
