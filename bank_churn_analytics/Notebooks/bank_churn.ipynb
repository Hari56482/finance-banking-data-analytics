```python
import pandas as pd

df = pd.read_csv("customers.csv")
df.head()
```

```python
df['churn'].value_counts()
```

```python
churn_rate = df['churn'].mean() * 100
print("Churn Rate: ", churn_rate, "%")
```

```python
summary = df.groupby('geography')['churn'].mean().reset_index()
summary['churn'] = summary['churn'] * 100
summary.to_csv("churn_summary_by_country.csv", index=False)
summary
```
